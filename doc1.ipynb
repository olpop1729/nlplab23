Frequency Distribution

```python
fdist1 = FreqDist(text1)
print(fdist1)
fdist1.plot(50)
fdist1.plot(50, cumulative=True)
```

```python
fdist1.most_common(20)
```

Fine-grained selection of words

```python
vocab = set(text1)
long_words = [word for word in vocab if len(word) > 15]
sorted(long_words)
```

```python
sorted(word for word in set(text1) if len(word) > 7 and fdist1[word] > 7)
```

Accessing Text Corporus

```python
import nltk
nltk.corpus.gutenberg.fileids()
```

Here gutenberg is the name of one of the corpus. The "fileids" function lists all the text files associated with that corpus.

```python
nltk.corpus.inaugural.fileids()
```

```python
emma = nltk.corpus.gutenberg.words('austen-emma.txt')
len(emma)
```

```python
emma = nltk.Text(nltk.corpus.gutenberg.words('austen-emma.txt'))
emma.concordance('surprize')
```
